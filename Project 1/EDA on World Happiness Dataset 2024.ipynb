# Exploratory Data Analysis

```python
#histogram of happiness score and Features
sns.histplot(df["Life Ladder"],kde="True")
plt.xlabel("Life ladder")
plt.ylabel("Count")
```

The distribution shows that majority of the countries fall in the range 4-7.
Given that the average is 5.4, it shows that majority of the countries are averagely happy.

```python
#correlation matrix
correlation_matrix=df.corr()
correlation_matrix
```

As per the correlation matrix, the main factors affecting Happinesss in various countries are Log GDP Per capita,Healthy Life expectany at birth and Social Support.This is because they have the highest correlation to Life ladder(Happiness score).
High log GDP and Social support means we get a higher happiness score.

On the other side a Perception of corruption score has the lowest correlation to ladder score. This means that a high Perception of corruption score leads to low score of happiness and vice versa.

Below are scatterplots to illustrate the same

```python
#scatterplots
plt.scatter(x=df["Log GDP per capita"],y=df["Life Ladder"])
plt.xlabel("Log GDP per capita")
plt.ylabel("Life Ladder")
plt.title("life ladder vs Log GDP per capita")
```

```python
plt.scatter(x=df["Social support"],y=df["Life Ladder"])
plt.xlabel("Social support")
plt.ylabel("Life Ladder")
plt.title("life ladder vs Social support")
```

```python
plt.scatter(x=df["Healthy life expectancy at birth"],y=df["Life Ladder"])
plt.xlabel("Healthy life expectancy at birth")
plt.ylabel("Life Ladder")
plt.title("life ladder vs Healthy life expectancy at birth")
```

```python
plt.scatter(x=df["Perceptions of corruption"],y=df["Life Ladder"])
plt.xlabel("Perceptions of corruption")
plt.ylabel("Life Ladder")
plt.title("life ladder vs Perceptions of corruption")
```
